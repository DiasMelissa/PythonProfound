# **Seaborn**

It used to create statistical graphs.

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
import seaborn as sea
```

```python
data = sea.load_dataset("tips")
```

```python
data.head()
```

```python
sea.jointplot("total_bill", "tip", data, kind = 'reg');
```

```python
sea.lmplot("total_bill", "tip", data, col = 'smoker');
```

```python
df= pd.DataFrame()
```

```python
df['a'] = random.sample(range(1, 200), 25)
df['b'] = random.sample(range(1, 200), 25)
```

```python
df.head()
```

```python
sea.lmplot('a', 'b', data = df, fit_reg = True)
```

```python
sea.kdeplot(df.b);
```

```python
sea.kdeplot(df.a, df.b);
```

```python
sea.distplot(df.a);
```

```python
plt.hist(df.a, alpha = 0.3)
sea.rugplot(df.a);
```

```python
sea.boxplot(x='day', y='tip', data=data);
```

```python
sea.violinplot(x = 'day', y= 'tip', data = data);
```

```python
sea.heatmap([df.a, df.b], annot = True, fmt = "d");
```

```python
sea.clustermap(df);
```
